# General plot.**py**

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```
x = np.random.random((10, 1))

print(x)
```

```
plt.plot(x, '*-')
plt.show()
```

```
x = np.linspace(0, 100, 50)
y = np.power(x, .5)
plt.plot(x, y, '+-')
plt.show()
```

```
sns.set()
sns.lineplot(x, y)
plt.show()
```

# **Plotting Data from Dataset**

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```
data = pd.read_csv('machine_learning/Machine_learning/3_Mobile_dataset/mobile_cleaned.csv')
data.head()
```

```
data.describe()
```

# Uni-Variant

```
sns.countplot(x ='sim_type', data = data)
plt.show()
```

```
sns.distplot(data["stand_by_time"])#maintained bet 200-400
plt.show()
```

```
sns.histplot(data=data, x="stand_by_time")#maintained between 200-400
plt.show()
```

# Bi-Variant

```
ax = sns.scatterplot(x="stand_by_time", y="battery_capacity", data=data)
plt.show()#helps to understand the correlation/outliers which might disintegrate our model reading/model struggles to remove these outliers
```

```
#Minimum, 25,50,75, max lines in the boxplots
ax = sns.boxplot(x="is_liked", y="battery_capacity",data=data)#used to understand the outliers more easily
plt.show()
#Minimum = Q1–1.5 * IQR and Maximum = Q3 + 1.5 * IQR
#the first line represents min value 2nd line 25%, 3rd line 50%, 4th line 75%,5th line 100% in the boxplot
#using box calc max n min ranges
```

```
ax = sns.boxplot(x = "expandable_memory", y = "price", data=data)
plt.show()
```

# Multi-Variant

```
ax = sns.violinplot(x='expandable_memory', y='price',z='battery_capacity', data=data,
              hue='is_liked', split=True)
plt.show()#if we want to consider 3data  we use violin plot
```

```
ax = sns.scatterplot(x = "stand_by_time", y = "battery_capacity", hue="thickness",data=data)
#hue is the additional parameter to use 4th paremeter we use scatter plots
plt.show()
```

```
dataframe = pd.read_csv("machine_learning/Machine_learning/2_IRIS/iris.csv")
print(dataframe.head())


ax = sns.scatterplot(x="petal.length", y="petal.width",hue="Variety",size = "sepal.width",data=dataframe)
plt.show()
```

```
ax = sns.swarmplot(x='sepal.length', y='Variety', data=dataframe)
#will plot every data and gives clear information 
```

```
sns.pairplot(dataframe[['petal.length','petal.width','sepal.length','sepal.width']])
plt.show()#comparisons within the data/ 1to1 analysis
```

```
sns.pairplot(dataframe, hue='Variety')#adding a parameter to the variety calculated previously
plt.show()
```
